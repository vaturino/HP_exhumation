```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dyn_name = "dynamic_ocean_sed_geometric_nosedcut_strongOPC_box100km_max7.5e22"
kin_name = "kinematic_basalt2.5_quartz1_resplugin_strongOPC"
dyn_loc = f"/home/vturino/PhD/projects/exhumation/plots/single_models/{dyn_name}/txt_files"
kin_loc = f"/home/vturino/PhD/projects/exhumation/plots/single_models/{kin_name}/txt_files"
dyn = pd.read_csv(f'{dyn_loc}/2D_v.txt', delim_whitespace=True)
kin = pd.read_csv(f'{kin_loc}/2D_v.txt', delim_whitespace=True)
kin.conv_rate[0]    = np.nan
```

```python
plt.plot(dyn.time/1e6, dyn.conv_rate, label='dynamic')
plt.plot(kin.time/1e6, kin.conv_rate, label='kinematic')
plt.xlabel('Time (Ma)')
plt.ylabel('Convergence rate (cm/yr)')
plt.legend()
plt.savefig(f"/home/vturino/PhD/projects/exhumation/plots/single_models/{kin_name}/compare_conv_rate.eps", dpi = 1000)
```
